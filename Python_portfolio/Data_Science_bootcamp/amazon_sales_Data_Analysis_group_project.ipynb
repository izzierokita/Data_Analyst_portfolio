### Conclusion

The results of the linear regression suggest that the explanatory variable 'verified_reviews' has little effect on the explanatory variable 'rating'. 
The coefficient values for the other explanatory variables are close to zero, meaning that they do not have a significant effect on 'rating.
 The intercept value is 4.26, which means that for samples for which all explanatory variables are 0, the expected value of the 
 'rating' is about 4.26. The R^2 ratio is -0.017, which means that the model does not explain a significant variance in the explanatory variable.
  The RMSE value is 0.31, which means that the average prediction error is about 0.31.Therefore, it can be concluded that the model is not a good 
  'rating' predictor for this dataset.

```python
model = DecisionTreeRegressor(
        max_depth=7,
        min_samples_leaf=5)

random_search = RandomizedSearchCV(model,
param_distributions={'max_depth': range(1,10),
                      'min_samples_leaf': range(1, 10)
                     },
n_iter=20,
cv=5,
n_jobs=1,
random_state=42
)


random_search.fit(X_train, y_train)

y_pred = random_search.best_estimator_.predict(X_val)

print("MAE Tree: ", mean_absolute_error(y_val,y_pred))
print("R2_ Tree: ", r2_score(y_val, y_pred))
print("RMSE Tree: ", sqrt(mean_squared_error(y_val, y_pred)))
```
